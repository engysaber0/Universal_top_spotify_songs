## Null check & Removing

# # Checking the outliers:

```python
# Getting Absolute of tempo
#we have 838 outliers in column tempo so we can drop it instead of replacing as it is un necessary column
df['tempo'] = np.abs(df['tempo'])
q1 = np.percentile(df['tempo'], 25)
q3 = np.percentile(df['tempo'], 75)
norm_range = (q3 - q1) * 1.5
lower_outliers = df[df['tempo'] < (q1 - norm_range)]
upper_outliers = df[df['tempo'] > (q3 + norm_range)]
outliers = len(lower_outliers)+len(upper_outliers)
print(f"The number of outliers in : {outliers}")
```

```python
df.drop(columns = ['tempo'], inplace = True)
```

# Con... data manipulation and data cleansing

```python
# Extract the year and create a new column
df['snapshot_year'] = df['snapshot_date'].dt.year

# Display the DataFrame with the new 'snapshot_year' column
print(df[['snapshot_date', 'snapshot_year']].head(10))
print(df[['snapshot_date', 'snapshot_year']].tail(10))
```

```python
# Create a copy of the DataFrame
df_copy = df.copy()

# Convert 'artists' column to string type in the copy
df_copy['artists'] = df_copy['artists'].astype(str)

# Group by 'artists' and calculate the average popularity
average_popularity_by_artist = df_copy.groupby('artists')['popularity'].mean()

# Display the result
print(average_popularity_by_artist)
```

```python
# Filter data to include only songs with a popularity score greater than 90
popular_songs = df[df['popularity'] > 90]

# Display the filtered DataFrame
print(popular_songs)
```

```python
# Sort the data based on 'daily_rank' column in ascending order, and then by 'popularity' column in descending order
sorted_data = df.sort_values(by=['daily_rank', 'popularity'], ascending=[True, False])

# Display the sorted DataFrame
print(sorted_data)
```

```python
# Group by 'country' and calculate the average popularity
average_popularity_by_country = df.groupby('country')['popularity'].mean()

# Display the result
print(average_popularity_by_country)
```
